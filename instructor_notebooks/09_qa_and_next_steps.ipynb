## 9.1 — The Identity Shift
**[Watch Only]**

> **Instructor note (3 min):** Read this section aloud or paraphrase it. This is the frame for everything that follows. The goal is not to make students feel good about what they built — it is to make them think differently about where the hard work actually is.

You ran a complete forecasting pipeline today. Six models, three evaluation stages, a defensible leaderboard, and a production recommendation.

That is genuinely useful. But here is what it is not:

**It is not a production system.**

What you built today is a proof of concept with a clean dataset, a locked config, precomputed artifacts, and no stakeholders. The gap between this notebook and a system your procurement team depends on at 6am is not a gap in model knowledge. It is a gap in systems thinking.

**The shift that separates practitioners from notebook runners:**

A notebook runner asks: *"Which model has the best wMAPE?"*  
A system architect asks: *"Which model can I deploy, monitor, debug, and explain — at the cadence my business requires, within the compute budget I have, with a team that can maintain it when I'm not in the room?"*

Those are different questions. The leaderboard answers the first one. **You have to answer the second one yourself.**

Today gave you the tools to ask it correctly.

- Start with `00_env_check.ipynb` on your own data. Validate schema, series count, history depth, and zero fraction before writing any model code.
- Lock a `config.py` equivalent before writing a single modeling cell. Every constant in one place, imported everywhere, never redefined.
- Run AutoETS on your data this week. That is your performance floor. Everything you build next gets compared to it.

**If you are evaluating forecasting vendors or tools:**
- Ask them for their evaluation methodology. If they report per-series averages instead of pooled metrics, discount the result.
- Ask them what their interval calibration approach is. If they do not have one, ask what they recommend for safety stock setting. The answer will tell you what you need to know.
- Use the rubric to evaluate their offering on Dimensions 3, 5, and 6. Most vendor pitches are strong on Dimension 1 (model accuracy) and silent on everything else.
