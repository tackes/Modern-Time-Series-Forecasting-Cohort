## 9.3 — The Companion Asset
**[Watch Only]**

The most common mistake practitioners make after a workshop like this is investing in the wrong thing next.

They add a seventh model when their monitoring is nonexistent. They tune hyperparameters when their feature pipeline has no backfill strategy. They optimize wMAPE when their intervals have never been evaluated.

**The Enterprise Forecasting Tech Stack Rubric** is a six-dimension maturity assessment that tells you where your stack actually is — and where to invest next.

It covers:
- Model selection discipline
- Feature pipeline architecture
- Serving and inference reliability
- Evaluation rigor
- Monitoring and drift detection
- Organizational ownership

Each dimension has a 1–4 scoring rubric with specific descriptors, a maturity band summary, and a gap analysis that tells you what to fix first.

**Find it in:** `companion_asset/enterprise_forecasting_tech_stack_rubric.md`

Score your current stack tonight. The lowest dimension is where to start.

## 9.4 — What to Do This Week
**[Watch Only]**

Pick one. Do it before next week.

**If you have a production forecasting system:**
- Score it against the rubric. Write down your lowest dimension. Calendar 2 hours to address it.
- Run the pooled wMAPE calculation on your existing models. Compare it to what you have been reporting. Investigate any gap.
- Add one precomputed baseline (SeasonalNaive) to your existing pipeline and score it against your current model. If the gap is under 3 wMAPE points, escalate the conversation about whether your ML infrastructure is justified.

**If you are building a forecasting system:**
- Start with `00_env_check.ipynb` on your own data. Validate schema, series count, history depth, and zero fraction before writing any model code.
- Lock a `config.py` equivalent before writing a single modeling cell. Every constant in one place, imported everywhere, never redefined.
- Run AutoETS on your data this week. That is your performance floor. Everything you build next gets compared to it.

**If you are evaluating forecasting vendors or tools:**
- Ask them for their evaluation methodology. If they report per-series averages instead of pooled metrics, discount the result.
- Ask them what their interval calibration approach is. If they do not have one, ask what they recommend for safety stock setting. The answer will tell you what you need to know.
- Use the rubric to evaluate their offering on Dimensions 3, 5, and 6. Most vendor pitches are strong on Dimension 1 (model accuracy) and silent on everything else.
